```python
import pandas as pd
from bs4 import BeautifulSoup

with open('data/example_review.html', 'r') as f:

    html_content = f.read() # Make sure you paste the given HTML string in place of [Your HTML code here].

soup = BeautifulSoup(html_content, 'html.parser')
player_rows = soup.select('tr.playerRow')

# Extracting data for each player
data = []
for row in player_rows:
    player_name = row.select_one('div.playerName').text
    player_details = row.select_one('div.playerDetails').text
    mins_played = row.select_one('td.xMins').text
    pts1 = row.select('td.ptsshoq')[0].text
    pts2 = row.select('td.ptsshoq')[1].text
    pts3 = row.select('td.ptsshoq')[2].text
    pts4 = row.select('td.blotti')[0].text
    pts5 = row.select('td.boldTD')[1].text  # Second td with boldTD class
    ownership = row.select_one('td.eliteOwn').text

    data.append([player_name, player_details, mins_played, pts1, pts2, pts3, pts4, pts5, ownership])

# Create DataFrame
df = pd.DataFrame(data, columns=['Player Name', 'Player Details', 'Minutes Played', 'Pts1', 'Pts2', 'Pts3', 'Pts4', 'Pts5', 'Ownership'])

print(df)
```
